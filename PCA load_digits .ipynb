```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
load_Data = load_digits()
load_Data.keys()
```

```python
print(load_Data.DESCR)
```

```python
df = pd.DataFrame(load_Data['data'],columns=load_Data['feature_names'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
trsn = scaler.fit_transform(df)
```

```python
trsn
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pca  = PCA(n_components=2)
```

```python
data_pc = pca.fit_transform(trsn)
```

```python
pca.explained_variance_
```

```python
data_pc
```

```python
plt.Figure(figsize=(8,6))
plt.scatter(data_pc[:,0],data_pc[:,1], c=load_Data['target'] , cmap='plasma')
plt.xlabel('first prinicipal')
plt.ylabel('secend principal')
```
